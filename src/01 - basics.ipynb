```python
from qiskit import QuantumCircuit
from qiskit.providers.aer import AerSimulator
```

```python
# Create quantum circuit with 3 qubits and 3 classical bits
qc = QuantumCircuit(3, 3)
qc.draw(output='mpl') # Returns the drawing of the circuit
```

```python
qc = QuantumCircuit(3, 3)
# Measure qubits 0, 1 and 2 to classical bits 0, 1 and 2 respectively
qc.measure([0, 1, 2], [0, 1, 2])
qc.draw(output='mpl')
```

```python
sim = AerSimulator() # Create a new simulator object
```

```python
job = sim.run(qc)      # Run the experiment
result = job.result()  # Get the results
result.get_counts()    # Interpret the results as a "counts" dictionary
```

```python
# Create quantum circuit with 3 qubits and 3 classical bits:
qc = QuantumCircuit(3, 3)
qc.x([0, 1]) # Perform X-gates on qubits 0 and 1
qc.measure([0, 1, 2], [0, 1, 2])
qc.draw(output='mpl')
```

```python
job = sim.run(qc)      # Run the experiment
result = job.result()  # Get the results
result.get_counts()    # Interpret the results as a "counts" dictionary
```

```python
# Create quantum circuit with 2 qubits and 2 classical bits:
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0, 1)                    # q1 = XOR(q0, q1)
qc.measure([0, 1], [0, 1])
display(qc.draw(output='mpl')) # Display the drawing of the circuit

job = sim.run(qc)
result = job.result()
result.get_counts()
```
